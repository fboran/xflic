```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks, peak_prominences
```

```python
df = pd.read_csv("../waveform/waveforms_ch0_run454.txt",
                 delim_whitespace=True,
                 usecols=[100,])
                 #names=["y",])
sig=df.to_numpy()
plt.plot(sig)
#pmt signal is negative 
```

```python
s= -1*(sig-903.4) # baseline and (-1)
#s
t=df.index.to_numpy()
#t
```

```python
dashes = [100, 10]  # 10 points on, 5 off, 100 on, 5 off

fig, ax = plt.subplots()
line1, = ax.plot(s, '-', linewidth=0.2, label='test')
line1.set_dashes(dashes)


ax.legend(loc='lower right')
plt.show()
```

```python
plt.plot(s)
plt.axis([np.amin(t), np.amax(t), 1 * np.min(s), 1.1 * np.max(s)])


dashes = [10, 5, 100, 5]  # 10 points on, 5 off, 100 on, 5 off

line1, = ax.plot(s, '-.', linewidth=2)
line1.set_dashes(dashes)


```

```python
plt.plot( s)
plt.axis([np.amin(t), np.amax(t), 1 * np.min(s), 1.1 * np.max(s)])

# this is an inset axes over the main axes

n, bins, patches = plt.hist(s, 20, density=True, facecolor='g', alpha=1)

plt.axes([.65, .6, .2, .2], facecolor='b')
n, bins, patches = plt.hist(s, 20, density=True)


plt.title('Probability')
plt.xticks([])
plt.yticks([])
plt.show()
```
